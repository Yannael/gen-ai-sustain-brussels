# üó£Ô∏è Converting Text to Speech with OpenAI

With OpenAI's text-to-speech (TTS) feature, you can turn any piece of text into spoken audio using different AI voices.

In this example, we use the **`gpt-4o-mini-tts`** model (the last one) to generate speech from text.


```python
from pathlib import Path

# Directly specify the filename
speech_file_path = Path("speech.mp3")

with client.audio.speech.with_streaming_response.create(
    model="gpt-4o-mini-tts",
    voice="fable",
    input="I'm here today to take a training course on how to use LLMs.",
    instructions="Use a professional and positive tone.",
) as response:
    response.stream_to_file(speech_file_path)

```

The `voice` parameter lets you choose **which AI voice** will speak the text when generating the audio.

OpenAI provides several different voices, each with a unique style and tone.

# üåç Multilingual Text-to-Speech

OpenAI's text-to-speech also works with **many other languages**, not just English.  
The model automatically detects the language based on your input text.

In the example below, we ask the model to speak in **French**:

```python
from pathlib import Path

# Directly specify the filename
speech_file_path = Path("speech_fr.mp3")

with client.audio.speech.with_streaming_response.create(
    model="gpt-4o-mini-tts",
    voice="coral",
    input="Je suis ici aujourd'hui pour suivre une formation sur l'utilisation desiugjhgjhs.",
    instructions="Utilisez un ton professionnel et positif.",
) as response:
    response.stream_to_file(speech_file_path)

```

```python
from IPython.display import Audio
Audio("speech_fr.mp3")
```

# üéôÔ∏è Transcribing Audio to Text (Speech-to-Text)

OpenAI also lets you convert **spoken audio into written text** ‚Äî this is called **speech-to-text** or **transcription**.

In this example, we take the audio file `speech.mp3` (generated earlier) and transcribe what it says:


```python
transcript = client.audio.transcriptions.create(
  model="gpt-4o-transcribe",
  file=open("speech.mp3", "rb")
)

transcript.text
```
